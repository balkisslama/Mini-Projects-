```python
def show_samples(X, y, class_names=None, nrows=2, ncols=6):
    fig, axes = plt.subplots(nrows, ncols, figsize=(10, 5))
    idIndex=np.arange(12)
    for i, ax in enumerate(axes.ravel()):
        ax.imshow(X[i])
        title = class_names[y[idIndex]]
        ax.set_title(title)
        ax.axis("off")
    plt.tight_layout()
    plt.show()
```

```python
def show_images (images,labels,class_names, n=12):
    plt.figure(figsize=(12,4))
    idxs=np.arange(n)
    for i ,idxs in enumerate(idxs):
        plt.subplot(2,n//2,i+1)
        plt.imshow(images[idxs])
        plt.title(class_names[labels[idxs]])
    plt.show()
```

```python
show_images(x_train_full,y_train_full,class_names,n=12)
```
